```python
# Importing the necessary Python libraries

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta
from typing import List
```

```python
# Part A:
class Mission:
    """
    A class to represent a space mission.
    """

    def __init__(self, name: str, launch_date: str):
        """
        Constructor to initialize the Mission object.

        Parameters:
        - name: Name of the mission
        - launch_date: Launch date in 'yyyy-mm-dd' format
        """
        self.name = name
        self.launch_date = launch_date
# End of class
```

```python
#Part A_cont.:

# Mission data for visualizations:

mercury = Mission("Mercury", "1962-02-20")

gemini = Mission("Gemini IV", "1965-06-03")

apollo = Mission("Apollo 11", "1969-07-16")


# Helper function to parse duration
def parse_duration(duration_str):
    duration = timedelta()
    parts = duration_str.split(", ")
    for part in parts:
        if "days" in part:
            days = int(part.split()[0])
            duration += timedelta(days=days)
        elif "hours" in part:
            hours = int(part.split()[0])
            duration += timedelta(hours=hours)
        elif "minutes" in part:
            minutes = int(part.split()[0])
            duration += timedelta(minutes=minutes)
    return duration

# # Function to visualize mission timeline
# def plot_mission_timeline(mission_objects):
#     fig, ax = plt.subplots(figsize=(10, 6))
#     mission_colors = {
#     "Friendship 7": "blue",
#     "Gemini IV": "green",
#     "Apollo 11": "red"
# }
    
#     for mission in mission_objects:
#         launch_datetime = datetime.strptime(f"{mission.launch_date} {mission.launch_time}", "%Y-%m-%d %H:%M %Z")
#         duration = parse_duration(mission.duration)
#         end_datetime = launch_datetime + duration
        
#         # Determine the mission type for coloring
#         mission_type = "Mercury" if "Mercury" in mission.name else "Gemini" if "Gemini" in mission.name else "Apollo"
        
#         # Plot the mission timeline
#         ax.plot([launch_datetime, end_datetime], [mission.name, mission.name], 
#         color=mission_colors[mission.name], linewidth=4, label=mission.name)

#         ax.plot(launch_datetime, mission.name, 'o', color=mission_colors[mission.name])
#         ax.text(launch_datetime, mission.name, " Launch", verticalalignment='center', fontsize=9)

#         ax.plot(end_datetime, mission.name, 'o', color=mission_colors[mission.name])
#         # ax.text(end_datetime, mission.name, " End", verticalalignment='center', fontsize=9)

    
#     # Formatting the plot
#     ax.set_title("Mission Timelines for Mercury, Gemini, and Apollo")
#     ax.set_xlabel("Timeline")
#     ax.set_ylabel("Missions")
#     ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d %H:%M"))
#     ax.xaxis.set_major_locator(mdates.DayLocator(interval=1))
#     plt.xticks(rotation=45)
#     plt.legend(loc="upper left")
#     plt.tight_layout()
#     plt.show()

#     for mission in mission_objects:
#         plt.plot(mission.launch_date, 0)


# # Plot the timelines
# plot_mission_timeline([mercury, gemini, apollo])
```

```python
#Part A_final.:

for mission in [mercury, gemini, apollo]:
    plt.plot(mission.launch_date, 0, 'o', label=mission.name)

plt.legend()
plt.gca().yaxis.set_ticks([]) 
plt.xlabel("Launch Date")
plt.title("Space Mission Launch Dates")
plt.show()
```

```python
# Part B:

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# I am defining a simplified flight path data for Apollo 13 (hypothetical coordinates)
# Crew of Apollo 8: Frank Borman, Jim Lovell, and Bill Anders
# Significane of Apollo 8: The crew of Apollo 8 were the first humans to orbit the Moon, the first to see its far side, and the first to broadcast live TV coverage from space. They also achieved the first crewed launch of the Saturn V rocket, became the first humans to travel into deep space, and took the first pictures of Earth from deep space.
flight_paths = {
    "Apollo 8": {
        "earth_orbit": [(0, 0, 0), (1, 1, 0), (1.5, 1.5, 0)],  # Points around Earth orbit
        "trans_lunar": [(1.5, 1.5, 0), (3, 3, 1.5)],           # Path to Moon
        "lunar_orbit": [(3, 3, 1.5), (3.5, 3.5, 1.5), (4, 4, 1.5)],  # Lunar orbit path
        "return_earth": [(4, 4, 1.5), (0, 0, 0)]                # Return path to Earth 
    }
}

# Plotting flight paths in 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")

# Loop through missions and plot each trajectory
for mission, paths in flight_paths.items():
    for segment, points in paths.items():
        points = np.array(points)
        ax.plot(points[:, 0], points[:, 1], points[:, 2], label=f"{mission} - {segment}")

# Add scatter for Earth (starting point)
ax.scatter(0, 0, 0, color="blue", s=200, label="Earth (Start)")  # Blue scatter for Earth

# Add scatter for Moon (ending destination)
moon_position = (4, 4, 1.5)
ax.scatter(*moon_position, color="grey", s=150, label="Moon (End)")  # Grey scatter for Moon

# Labels and customization
ax.set_title("Flight Plan Trajectories for Apollo 8")
ax.set_xlabel("X Distance (approx.)")
ax.set_ylabel("Y Distance (approx.)")
ax.set_zlabel("Z Distance (approx.)")
ax.legend()

plt.show()
```
